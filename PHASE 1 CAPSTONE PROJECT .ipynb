```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/nishusharma1608/India-Census-2011-Analysis/master/india-districts-census-2011.csv')
```

```python
print(data.columns)
```

```python
state_population = data.groupby('State name')['Population'].sum()
```

```python
print("Population of each state:")
print(state_population)
```

```python
state_population.plot(kind='bar', figsize=(10, 6), title='Population of each state')
plt.xlabel('State')
plt.ylabel('Population')
plt.show()
```

```python
state_literacy_rate = data.groupby('State name')['Literate'].sum() / data.groupby('State name')['Population'].sum() * 100
```

```python
print("Literacy rate in each state:")
print(state_literacy_rate)
```

```python

state_literacy_rate.plot(kind='bar', figsize=(10, 6), title='Literacy rate in each state')
plt.xlabel('State')
plt.ylabel('Literacy Rate')
plt.show()
```

```python
data = pd.read_html('https://statisticstimes.com/geography/india/indian-states-area.php')
```

```python
state_area_data = data[1]
```

```python
print(state_area_data.columns)
```

```python
state_area_data = state_area_data.rename(columns={'State/UT': 'State', 'Area(sq. km)': 'Area'})
```

```python
print("Area of each state:")
print(state_area_data)
```

```python
state_area_data.plot(kind='bar', figsize=(10, 6), legend=False)
plt.title('Area of Each State')
plt.xlabel('State')
plt.ylabel('Area (sq. km)')
plt.xticks(rotation=90)
plt.show()
```
